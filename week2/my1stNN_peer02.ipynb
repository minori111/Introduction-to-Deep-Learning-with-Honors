```python
from preprocessed_mnist import load_dataset
X_train, y_train, X_val, y_val, X_test, y_test = load_dataset()
print(X_train.shape, y_train.shape)
import matplotlib.pyplot as plt
%matplotlib inline
plt.imshow(X_train[0], cmap="Greys");
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X_train = X_train.reshape(X_train.shape[0], -1).astype('float32')
X_val = X_val.reshape(X_val.shape[0], -1).astype('float32')
X_test = X_test.reshape(X_test.shape[0], -1).astype('float32')

m, n = X_train.shape
n_classes = len(set(y_train))

sess = tf.Session()
y_train = sess.run(tf.one_hot(y_train, n_classes))
y_val = sess.run(tf.one_hot(y_val, n_classes))
y_test = sess.run(tf.one_hot(y_test, n_classes))
sess.close()
```

```python
def forward_propagation(X, w1, b1, w2, b2):
    z1 = tf.add(tf.matmul(X, w1), b1)
    a1 = tf.sigmoid(z1)
    z2 = tf.add(tf.matmul(a1, w2), b2)
    return z2
```

```python
def loss(z, Y, w1, w2, lamda = 0):
    return tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=z, labels=Y)) \
    + 0.5 * lamda * (tf.nn.l2_loss(w1) + tf.nn.l2_loss(w2))
```

```python
import time
start = time.time()
alpha = 0.01
lamda = 0.001
layer1 = 80
layer2 = 10
epochs = 500

tf.set_random_seed(777)
x = tf.placeholder(tf.float32, [None, n])
y = tf.placeholder(tf.float32, [None, n_classes])
w1 = tf.Variable(initial_value=tf.random_normal([n, layer1]), dtype=tf.float32)
w2 = tf.Variable(initial_value=tf.random_normal([layer1, n_classes]), dtype=tf.float32)
b1 = tf.Variable(initial_value=tf.random_normal([1, layer1]), dtype=tf.float32)
b2 = tf.Variable(initial_value=tf.random_normal([1, n_classes]), dtype=tf.float32)
z = forward_propagation(x, w1, b1, w2, b2)
cost = loss(z, y, w1, w2, lamda)
accuracy = 100 * tf.reduce_mean(tf.cast(tf.equal(tf.argmax(z, axis=1), tf.argmax(y, axis=1)), tf.float32))

optimizer = tf.train.AdamOptimizer(learning_rate = alpha, beta1=0.9, beta2=0.999).minimize(cost)

sess = tf.Session()
sess.run(tf.global_variables_initializer())

j_history = []
j_val_history = []

for epoch in range(epochs):
    _, j, acc = sess.run([optimizer, cost, accuracy], feed_dict={x: X_train, y: y_train})
    j_val, acc_val = sess.run([cost, accuracy], feed_dict={x: X_val, y: y_val})
    
    j_history.append(j)
    j_val_history.append(j_val)
    print("Epoch: %d\tCost: %0.3f\tTrain Accuracy: %0.2f\tValidation Accuracy: %0.2f" % (epoch + 1, j, acc, acc_val))
    
train, = plt.plot(range(1, epochs + 1), j_history)
val, = plt.plot(range(1, epochs + 1), j_val_history)
plt.xlabel("No. of Iterations")
plt.ylabel("Cost")
plt.legend([train, val], ['Train Cost', 'Validation Cost'])
plt.show()

print("Training accuracy: %0.2f\tTest set accuracy: %0.2f\tValidation set accuracy: %0.2f"
      % (sess.run(accuracy, feed_dict={x: X_train, y: y_train}), 
        sess.run(accuracy, feed_dict={x: X_test, y: y_test}), 
        sess.run(accuracy, feed_dict={x: X_val, y: y_val})))
end = time.time()
print("Execution time: %0.2f s" % (end - start))
```

```python
print("Test Accuracy: %0.2f %%" % (sess.run(accuracy, feed_dict={x: X_test, y: y_test})))
```
